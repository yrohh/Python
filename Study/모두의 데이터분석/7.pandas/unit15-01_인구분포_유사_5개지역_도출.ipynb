# 알고리즘

- 데이터 로드
- 모든 연령값 행별 총인구수로 나눠 비율 산출
- 필요없는 열 제거
- 기준 지역 입력
- 입력된 지역명을 포함하는 인덱스의 행 도출
- 모든 지역별로 기준 지역의 값을 뺀 뒤, 제곱
- 제곱한 값들을 합함
- 합한 제곱값들의 크기를 기준으로 오름차순하여 5번째 행까지 도출
- 행과 열 치환하여 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family ='Malgun Gothic')
df = pd.read_csv('age.csv', encoding='cp949', index_col = 0) 
df = df.div(df['총인구수'], axis = 0)     # ①-1. 전체 데이터를 총인구수로 나눠서 비율로 변환
del df['총인구수'], df['연령구간인구수']  # ①-2. 총인구수, 연령구간인구수 열 삭제
# name = input('원하는 지역의 이름을 입력해주세요 : ')  # ② 지역 이름 입력
a = df.index.str.contains("신도림동")  # ③ 해당 행을 찾아서 해당 지역의 인구 구조를 저장
df2 = df[a]
df2
b = df.sub(df2.iloc[0],axis=1) # 데이터프레임 간 연산은 안되므로, series를 도출하여 연산.
c = np.power(b,2) # numpy의 제곱 함수.
d = c.sum(axis=1)
e = d.sort_values(ascending=True).index[:5] # 오름차순 정렬 후 상위 5개 인덱스 도출. 인덱스명이 도출되므로, loc를 통해 접근해야 함.
df.loc[e].T.plot()
```

### 코드 간략화.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family ='Malgun Gothic')
df = pd.read_csv('age.csv', encoding='cp949', index_col = 0) 
df = df.div(df['총인구수'], axis = 0)     # ①-1. 전체 데이터를 총인구수로 나눠서 비율로 변환.
del df['총인구수'], df['연령구간인구수']  # ①-2. 총인구수, 연령구간인구수 열 삭제.
name = input('원하는 지역의 이름을 입력해주세요 : ')  # ② 지역 이름 입력.
a = df.index.str.contains(name)  # ③ 해당 행을 찾아서 해당 지역의 인구 구조를 저장.
df2 = df[a]
df.loc[np.power(df.sub(df2.iloc[0], axis = 1), 2).sum(axis = 1).sort_values().index[:5]].T.plot()
plt.show()
```
