```python
import csv
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt

f = open('age.csv')
data = csv.reader(f)


def make_g(locate) :
    result = []
    for row in data:
        if locate in row[0]:
            for i in row[3:]:
                result.append(int(i))
    
    
    plt.style.use('seaborn')
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize = (10,5), dpi=300)    
    plt.barh(range(101), result)
    plt.title(locate + ' 지역의 인구 구조')
    plt.show()
    
    
lo = input('지역명 입력 : ')
make_g(lo)
```

```python
import csv
import pandas as pd
from pandas import DataFrame as df
import matplotlib.pyplot as plt

f = open('age2.csv')
data = csv.reader(f)


def make_g(locate) :
    m = []
    f = []
    for row in data:
        if locate in row[0]:
            for i in row[106:207]:
                m.append(-int(i))
            for j in row[209:]:
                f.append(int(j))
    
    
    # plt.style.use('seaborn')
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(10,5), dpi=300)
    plt.title(locate + ' 지역의 인구 구조')
    plt.barh(range(101), m, label='남성', color='blue')
    plt.barh(range(101), f, label='여성', color='red')
    plt.legend()
    plt.show()
    
    
lo = input('지역명 입력 컴온')
make_g(lo)
```
