# 2.4 그래프 작성

```python
# -*- coding: utf-8 -*-
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

FLAG_fig = False
```

```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
x
```

# 2.4.2 복수의 그래프
하나의 배치에 여러 개의 그래프 출력 

```python
fig = plt.subplots(figsize=(10,3)) # size [inch,inch]
plt.plot(x,y1, label="red", color='red')
plt.plot(x,y2, label="blue", color= 'blue')
plt.plot(x,y3, label="orange", color='orange')
plt.plot(x,y4, label="green", color='green')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example')
plt.legend()

plt.tight_layout() # xlabel이 누락되는 것을 방지하는 방법１
if FLAG_fig: plt.savefig('fig_PLT_MultipleGraph_1.png')
#if FLAG_fig: plt.savefig('fig_PLT_MultipleGraph_1.png', bbox_inches='tight') #xlabe이 누락되는 것을 방지하는 방법 2
plt.show()
```

### 서로 다른 좌표에 그래프 출력

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))
 
axs[0, 0].plot(x, y1) # upper left
axs[0, 1].plot(x, y2) # upper right
axs[1, 0].plot(x, y3) # lower left
axs[1, 1].plot(x, y4) # lower right

if FLAG_fig: plt.savefig('fig_PLT_MultipleGraph_2.png')
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(6,4))

ax1.plot(x, y1) # upper left
ax2.plot(x, y2) # upper right
ax3.plot(x, y3) # lower left
ax4.plot(x, y4) # lower right
 
if FLAG_fig: plt.savefig('fig_PLT_MultipleGraph_3.png')
plt.show()
```

# 2.4.3 Titanic의 pandas 그래프 그리기

```python
# -*- coding: utf-8 -*-
import numpy as np
import pandas as pd
#import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline  

FLAG_fig = False
```

```python
titanic_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv" #트레이닝 데이터를 읽어들인다
df = pd.read_csv(titanic_url) # df; DataFrame의 약어
# df.to_csv('titanic_train.csv') # 읽어들인 데이터의 출력
df.head()
```

```python
print(' type(df) is ',type(df),'\n','len(df) =',len(df))
```

```python
df['Age'].fillna(df.Age.median(), inplace=True) #inplace=True 는 처리 경감을 위해, 원래의 데이터를 처리한다
```

```python
hist = df['Age'].hist(bins=16)
# matplotlib를 이용한 plt.hist(df['Age'],  bins=16)와는 거의 동일

if FLAG_fig: plt.savefig('fig_PLT_Titanic_hist_01.png')
plt.show()
```

```python
age1 = (df['Age'] < 15).sum()
age2 = ( (df['Age'] >= 15) & (df['Age'] < 60) ).sum()
age3 = (df['Age'] >= 60).sum()
print(age1, age2, age3, (age1+age2+age3))
```

```python
series2 = pd.Series([age1, age2, age3], index=['Child', 'Adult', 'Elderly'], name='Age')
series2.plot.pie(figsize=(4, 4), autopct='%.1f%%')
plt.legend('upper right')
plt.tight_layout() # xlabel이 누락되는 것을 방지하는 방법１
if FLAG_fig: plt.savefig('fig_PLT_Titanic_pie_01.png')
plt.show()
```

```python
cross_01 = df.pivot_table(index=['Survived'], columns=['Sex'],  \
                          values=['PassengerId'],aggfunc='count', fill_value=0)
print(cross_01)
```

```python
cross_01.plot(kind='bar')

if FLAG_fig: plt.savefig('fig_PLT_Titanic_cross_01_hist.png')
plt.show()
```

sns.set()는，Seaborn의 디폴트 스타일을 적용한다<br>
https://seaborn.pydata.org/generated/seaborn.set.html<br>
set 함수는 스타일, 팔레트, 글꼴, 글꼴 크기 등을 설정. 이는 앞으로 그리게 되는 모든 플롯의 그림에 영향을 준다. 즉, 환경변수 설정이다. 

```python
sns.set()
# Load the example tips dataset
iris = sns.load_dataset("iris") # type (iris) = pandas.core.frame.DataFrame
iris.head()
# iris.info() 데이터프레임
```

lmplot()는，회귀모델의 생성과 그래프 작성을 동시에 수행한다<br>
다음 예는 가로축을 sepal_length[cm], 세로축을 sepal_width[cm]로 하고, 3종류의 iris 각각에 대해 회귀모델을 고려한다<br>
truncate=False는 각 회귀모델을 가로축의 영역에 직선을 그린다. =True는，데이터의 범위 안으로만 한다<br>
size=5 는 그래프의 크기[inch], hue는 색조를 지정<br>
다만，통계량은 출력하지 않는다<br>
참조：<br>
http://seaborn.pydata.org/generated/seaborn.lmplot.html <br>
https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python
# Plot tip as a function of toal bill across days
# size is for old version, height is for new version.

g = sns.lmplot(x="sepal_length", y="sepal_width", hue="species",
               truncate=True, height=4, data=iris)
# 혹시、FutureWaring이 표시되어도 무시해도 상관없다.

# Use more informative axis labels than are provided by default
g.set_axis_labels("Sepal length (cm)", "Sepal width (cm)")

if FLAG_fig: g.savefig('fig_PLT_Iris_sns_regression.png')
plt.tight_layout() # xlabel이 누락되는 것을 방지하는 방법１
```

```python
g = sns.lmplot(x="sepal_length", y="sepal_width", hue="species",
               truncate=False, height=4, data=iris)
# 혹시、FutureWaring이 표시되어도 무시해도 상관없다.

# Use more informative axis labels than are provided by default
g.set_axis_labels("Sepal length (cm)", "Sepal width (cm)")
plt.tight_layout() # xlabel이 누락되는 것을 방지하는 방법１
```

### 산점도 행렬 pairplot()

pairplot()은 각 변수가 조합된 산점도, 즉<br>
산점도 행렬을 그린다.<br>
참조<br>
https://seaborn.pydata.org/generated/seaborn.pairplot.html <br>
인수 diag_kind='kde'; kernel density estimate (default),  ='hist'；histogram

```python
sns_plot = sns.pairplot(iris, hue="species", diag_kind='hist')
plt.tight_layout() # xlabel이 누락되는 것을 방지하는 방법１
if FLAG_fig: sns_plot.savefig('fig_PLT_Iris_sns_pairplot.png')
```
